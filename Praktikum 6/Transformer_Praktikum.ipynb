```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
# 1. Load tokenizer dan model pretrained
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

```

```python
# 2. Tokenisasi kalimat
text = ["Deep learning with Transformers is powerful",
        "Transformers make deep learning models very strong"]

inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
```

```python
# 3. Forward pass ke model
outputs = model(**inputs)

```

```python
# 4. Ekstraksi embedding
last_hidden_states = outputs.last_hidden_state

print("Shape embedding:", last_hidden_states.shape)
```

```python
# 5. Visualisasi sederhana token dan vektornya

print("\nToken dan panjang embedding untuk setiap kalimat:")
for sentence_idx, sentence in enumerate(text):
    print(f"\nKalimat {sentence_idx + 1}: \"{sentence}\"")

    # Get tokens by decoding the input_ids for the current sentence
    # This will include special tokens like [CLS], [SEP], [PAD]
    sentence_input_ids = inputs['input_ids'][sentence_idx]
    tokens_for_sentence = tokenizer.convert_ids_to_tokens(sentence_input_ids)

    # The last_hidden_states shape is [batch_size, sequence_length, hidden_size]
    # We use the attention_mask to get the actual length of the sentence (excluding padding)
    actual_length = int(inputs['attention_mask'][sentence_idx].sum())

    for i in range(actual_length):
        token = tokens_for_sentence[i]
        embedding = last_hidden_states[sentence_idx][i][:5]  # take the first 5 dimensions
        print(f"{token}: {embedding}")
```

```python
import torch.nn.functional as F

# Extract [CLS] token embeddings for each sentence
# The [CLS] token is always at index 0 for BERT outputs
sentence_embedding_1 = last_hidden_states[0, 0, :].unsqueeze(0)  # Shape: [1, hidden_size]
sentence_embedding_2 = last_hidden_states[1, 0, :].unsqueeze(0)  # Shape: [1, hidden_size]

# Calculate Cosine Similarity
cosine_similarity = F.cosine_similarity(sentence_embedding_1, sentence_embedding_2)

print(f"Embedding kalimat 1 (CLS token): {sentence_embedding_1.shape}")
print(f"Embedding kalimat 2 (CLS token): {sentence_embedding_2.shape}")
print(f"Cosine Similarity antara kedua kalimat: {cosine_similarity.item()}")
```
